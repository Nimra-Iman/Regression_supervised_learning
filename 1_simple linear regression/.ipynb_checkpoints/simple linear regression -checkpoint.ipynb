<h3> practical implementation </h3>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dataset=pd.read_csv("placement.csv")
dataset.head()

# dataset m target continuos ka h is liye regression, regression m linear is liye k x and y m relatioship h
# or linear regression m simple linear regression is liye  input ek hi feature h
```

```python
dataset.isnull().sum()
```

```python
# first check relationship k regression algo m kon ca algo lgy ga
plt.figure(figsize=(4,4))
sns.scatterplot(x="cgpa",y="package", data=dataset)
plt.show()     #yani they have linear relationship so simple linear algo will be used
```

```python
x=dataset["cgpa"]
x.ndim #output=1
x=dataset[["cgpa"]]
x.ndim #output=2
                                # ML k algos ya sklearn s koi algo use kia to vo input features ko 2D array
                                # k tor pr hi accept krty hn, is liye is
                                # ko 2D array m convert kia
y=dataset["package"]
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
 # yani linear regression y=mx+c ka formula
# deta h , jis m unknown x and c hota h, y given data pr train ho ka x or c ki best value nikalta h, phir jab hm 
# is ko test krny k liye koi value pass kryn gy to y un calculated m and c ko formula m put kr k
# value predict kry ga, like simple example in below cell:
```

```python
lr.coef_   #yani m ki kia value derive ki is n training data ki madad s: 0.57018094
# y positive is liye ai q k grph line is below 90 degree
```

```python
lr.intercept_ #yani c ki kia value derive ki is n training data ki madad s: -0.980883767663578
# y negative is liye ai q k graph ki line negative y_axis ko touch kr rhi
```

```python
# so formula he calculated is: y=0.57018094*x + (-0.980883767663578)
# when we put x as 6.89, the calculated value is 2.94766291 which is same as calculated
# from this formula:
0.57018094* 6.89 + (-0.980883767663578)
```

```python
lr.predict(x_test)
```

```python
# 6.89	3.26
# simple

lr.predict([[6.89]])    #is n 2.94 dia jab k actually 3.26 tha yani thora qareeb dia, is liye han score
# check krty hn
```

```python
lr.score(x_test, y_test)
# gives 75% accuracy, which is not good one
```

```python
# jesy m n sab s ooper btaya k vo ek predicted line bnay ga or jab koi nia data ay ga to vo is 
# line m dekhy ga k kha pr is ka new data vali value ka result lie kry ga: to y predicted line esy show ho gi:
plt.figure(figsize=(3,2))
sns.scatterplot(x="cgpa", y="package", data=dataset)
predct=lr.predict(x)
#yhan hm lr.predict m x_test nhi daal skty
# q k dataset["cgpa"] and x_train ki no. of rows different hn, hm yhan x ko daaly gy
plt.plot(dataset["cgpa"],predct , color="red")
plt.legend(["original","predicted"])
plt.show()
```
